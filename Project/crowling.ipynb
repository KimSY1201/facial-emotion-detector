```python
from icrawler.builtin import GoogleImageCrawler, BaiduImageCrawler, BingImageCrawler

def download_images(flower_name, folder_name, max_num=100):
    crawler = BingImageCrawler(storage={'root_dir': folder_name})
    crawler.crawl(keyword=flower_name + ' 꽃', max_num=max_num)
```

```python
download_images('진달래', 'data/보라/진달래꽃')
download_images('진달래꽃', 'data/보라/진달래꽃')
download_images('꽃 진달래', 'data/보라/진달래꽃')
download_images('보라 진달래', 'data/보라/진달래꽃')
download_images('진달래꽃 보라', 'data/보라/진달래꽃')
```
